```python
# Cell 4 — Functions
# An agent is just a function that takes input and returns output

def classify_disruption(flight_id, delay_minutes, is_cancelled):
    """
    Detection Agent core logic.
    Takes flight data → returns disruption classification.
    """
    if is_cancelled:
        return {
            "flight_id":       flight_id,
            "disruption_type": "CANCELLATION",
            "severity":        5
        }
    elif delay_minutes >= 120:
        return {
            "flight_id":       flight_id,
            "disruption_type": "DELAY_CRITICAL",
            "severity":        3
        }
    elif delay_minutes >= 60:
        return {
            "flight_id":       flight_id,
            "disruption_type": "DELAY_MAJOR",
            "severity":        2
        }
    elif delay_minutes >= 15:
        return {
            "flight_id":       flight_id,
            "disruption_type": "DELAY_MINOR",
            "severity":        1
        }
    else:
        return {
            "flight_id":       flight_id,
            "disruption_type": "ON_TIME",
            "severity":        0
        }

# Test it with three scenarios
print(classify_disruption("DL-204", 75,  False))
print(classify_disruption("DL-317", 145, False))
print(classify_disruption("DL-408", 0,   True))
```

```python
# Cell 5 — Processing a full flight list
# The Detection Agent does not check one flight — it checks ALL of them

flights = [
    {"flight_id": "DL-101", "delay_minutes": 0,   "is_cancelled": False},
    {"flight_id": "DL-204", "delay_minutes": 75,  "is_cancelled": False},
    {"flight_id": "DL-317", "delay_minutes": 145, "is_cancelled": False},
    {"flight_id": "DL-408", "delay_minutes": 0,   "is_cancelled": True},
    {"flight_id": "DL-512", "delay_minutes": 22,  "is_cancelled": False},
]

print("--- Disruption Scan Report ---")
disruptions_found = []

for flight in flights:
    result = classify_disruption(
        flight["flight_id"],
        flight["delay_minutes"],
        flight["is_cancelled"]
    )
    if result["severity"] > 0:
        disruptions_found.append(result)
        print(f"⚠ {result['flight_id']}: {result['disruption_type']} "
              f"(severity {result['severity']})")

print(f"\nTotal disruptions detected: {len(disruptions_found)}")
print(f"Flights scanned: {len(flights)}")
```

```python
# Cell 6 — The Agent State
# In LangGraph, this dictionary is passed between ALL agents
# Each agent reads from it and adds its results back into it

# Initial state — what the system knows at start
state = {
    "flights_scanned":    0,
    "disruption_events":  [],
    "last_scan_time":     None,
    "status":             "idle"
}

print("--- Initial State ---")
print(state)

# Simulate the Detection Agent updating state
state["flights_scanned"]   = 5
state["disruption_events"] = disruptions_found   # from Cell 5
state["last_scan_time"]    = "2024-07-19 14:32:00"
state["status"]            = "disruptions_detected"

print("\n--- State After Detection Agent ---")
for key, value in state.items():
    print(f"  {key}: {value}")

print(f"\nSeverity summary:")
for event in state["disruption_events"]:
    print(f"  {event['flight_id']} → severity {event['severity']}")
```
